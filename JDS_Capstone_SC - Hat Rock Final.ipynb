### 3.1.1 Insight

```python
# Define columns where binary order [0, 1] applies
binary_order_columns = ['Potential_Customer', 'Status_Latest_Ad']
binary_order = [0, 1]  # Adjust as needed for your specific use case

# Mapping for specific column 'Status_Cust'
status_cust_labels = {
    'A': 'A = Active Buyer',
    'E': 'E = Inactive Buyer',
    'F': 'F = First Time Buyer',
    'L': 'L = Lapsing Buyer',
    'N': 'N = New Buyer',
    'S': 'S = Star Buyer'
}

# Create a grid of subplots
n_cols = 2  # Number of columns in the grid
n_rows = (len(CatCols) + n_cols - 1) // n_cols  # Calculate number of rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5))
axes = axes.flatten()  # Flatten axes for easier indexing

# Loop through each column and plot count plot
for i, col in enumerate(CatCols):
    if col in df.columns:  # Check if the column exists in the DataFrame
        ax = axes[i]

        # Determine the order
        if col in binary_order_columns:  # Apply binary order for specific columns
            order = binary_order
        else:  # For other categorical columns
            order = df[col].value_counts().index

        # Calculate proportions
        proportions = (df[col].value_counts(normalize=True) * 100).reindex(order)

        # Create the count plot
        sns.countplot(data=df, x=col, order=order, ax=ax, palette='viridis')
        ax.set_title(f'Count Plot for {col}', fontsize=14)
        ax.set_xlabel(col, fontsize=12)
        ax.set_ylabel('Count', fontsize=12)

        # Add percentage annotations on the bars
        for p, percentage in zip(ax.patches, proportions):
            ax.annotate(f'{percentage:.1f}%',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', fontsize=10, color='black',
                        xytext=(0, 5), textcoords='offset points')

        # Add legend with custom labels for 'Status_Cust'
        if col == 'Status_Cust':
            handles = []
            labels = []
            unique_categories = df[col].dropna().unique()
            for category in unique_categories:
                if category in status_cust_labels:
                    color = sns.color_palette('viridis', len(unique_categories))[list(unique_categories).index(category)]
                    handles.append(plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10))
                    labels.append(status_cust_labels[category])
            ax.legend(handles, labels, title=f'{col} Categories', loc='upper right')
    else:
        print(f"Column {col} does not exist in the DataFrame.")

# Hide any unused axes in the grid
for j in range(len(CatCols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway

f_group = df[df['Gender'] == 'F']['Cust_Ann_Income'].dropna()
m_group = df[df['Gender'] == 'M']['Cust_Ann_Income'].dropna()
u_group = df[df['Gender'] == 'U']['Cust_Ann_Income'].dropna()

f_stat, p_value = f_oneway(f_group, m_group, u_group)
print(f"ANOVA F-Statistic: {f_stat}, P-Value: {p_value}")
print(p_value<0.05)
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Prepare data for Tukey's test
data = df[['Gender', 'Cust_Last_Purchase']].dropna()

# Run Tukey's test
tukey = pairwise_tukeyhsd(endog=data['Cust_Last_Purchase'],  # Dependent variable
                          groups=data['Gender'],             # Grouping variable
                          alpha=0.05)                        # Significance level

# Print summary
print(tukey)

## Reject: Whether the null hypothesis is rejected (i.e., whether the difference is statistically significant). > SIGNIFICANT DIFFERENCE
```

```python
# Transformed training and testing data
X_train_pca
X_test_pca

ytrain
ytest

# Initialize models
svc = SVC()
lr = LogisticRegression()
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
rfc = RandomForestClassifier()

# List of models
models = [svc, lr, knn, dt, rfc]

# Parameter grid for SVC
svc_param_grid = {
    'kernel': ['linear'],
    'C': np.logspace(-2, 2, 10)
}

# Parameter grid for KNeighborsClassifier
knn_param_grid = {
    'n_neighbors': np.arange(1, 30),
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

# Parameter grid for LogisticRegression
lr_param_grid = {
    'C': np.logspace(-3, 3, 10),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear'],
    'fit_intercept': [True, False]
}

# Parameter grid for DecisionTreeClassifier
dt_param_grid = {
    'max_depth': np.arange(1, 20),
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 8],
    'max_features': ['log2']
}

# Parameter grid for RandomForestClassifier
rfc_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# List of parameter grids
params = [svc_param_grid, lr_param_grid, knn_param_grid, dt_param_grid, rfc_param_grid]
```

```python
def model_finder(train, test,y_train,y_test,model,params):
  # Initialize GridSearchCV with the model and parameters
  grid = GridSearchCV(estimator=model,
                  param_grid=params,
                  cv=3,
                  scoring = 'recall',
                  return_train_score=True)

  # Fit the grid search to the training data
  grid.fit(train,y_train)

  # Extract the best model, parameters, and score
  best_model = grid.best_estimator_
  best_par = grid.best_params_
  best_sc = grid.best_score_

  # Fit the best model to the training data
  best_model.fit(train,y_train)

  # Predict on the test data
  pred = best_model.predict(test)

  # Calculate evaluation metrics
  f1 = f1_score(y_test, pred, average='macro')
  conf_matrix = confusion_matrix(y_test,pred)
  report = classification_report(y_test,pred)
  recall = recall_score(y_test,pred)
  precision = precision_score(y_test,pred)

  # Return the best parameters, model, score, and evaluation metrics
  return best_par, best_model, best_sc, f1, recall, precision, conf_matrix, report
```

```python
# Initialize an empty DataFrame to store results
results = pd.DataFrame({'model': [], 'best_m': [], 'best_p': [], 'best_sc': [], 'F1': [], 'Recall':[], 'Precision': []})

# Loop through each model and its corresponding parameter grid
i=0
for m, p in zip(models, params):
  results.loc[i, 'model'] = str(m)
  try:
    # Get the best model, parameters, and evaluation metrics
    best_par, best_model, best_sc, f1, recall, precision, conf_matrix, report = model_finder(X_train_pca, X_test_pca, ytrain, ytest, m, p)

    # Store the results in the DataFrame
    results.loc[i, 'best_m'] = str(best_model)
    results.loc[i, 'best_p'] = str(best_par)
    results.loc[i, 'best_sc'] = round(best_sc, 2)
    results.loc[i, 'F1'] = round(f1, 2)
    results.loc[i, 'Recall'] = round(recall, 2)
    results.loc[i, 'Precision'] = round(precision, 2)

    # Print the results
    print(f'''\n
best model: {best_model}
best param: {best_par}
best score = {best_sc}
F1 = {f1}
confusion matrix:\n{conf_matrix}
classification report: \n{report}
    ''')
    print('\n***********************************\n')

  except Exception as e:
    print(f"Error processing {str(m)}: {str(e)}")

  i+=1
```

**Hint:**
    To calculate the revenue of a model for this new set of customers with different purchase rate we need to calculate the new confusion matrix given Precision and Recall for Class1 are fixed.

To calculate a **new confusion matrix** given a new purchase rate while keeping **precision** and **recall** fixed, follow these steps:

---

### **1. Definitions:**
- **Precision**: $ ( \text{Precision} = \frac{TP}{TP + FP} ) $
- **Recall**: $( \text{Recall} = \frac{TP}{TP + FN} ) $
- **Purchase Rate**: The proportion of positive cases in the dataset, $( \text{Purchase Rate} = \frac{TP + FN}{N} )$, where $( N )$ is the total number of customers.

---

### **2. Data:**

1. **Total Customers** = 30,000
2. **Purchase Rate** = 0.1
3. **Actual Purchase** = 0.1 * 30,000 = 3,000
4. **Non-Purchases** = 30,000 - 3,000 = 27,000

```python
customers = 30000
purchase_rate = 0.1
```

```python
def calculate_new_confusion_matrix(precision, recall, purchase_rate, total_samples):
    # Calculate the total number of positive samples (TP + FN)
    total_positives = purchase_rate * total_samples

    # Calculate True Positives (TP) using recall
    new_TP = recall * total_positives

    # Calculate False Positives (FP) using precision
    new_FP = new_TP / precision - new_TP

    # Calculate False Negatives (FN)
    new_FN = total_positives - new_TP

    # Calculate True Negatives (TN)
    new_TN = total_samples - (new_TP + new_FP + new_FN)

    # Return confusion matrix as integers
    return np.round([[new_TN, new_FP], [new_FN, new_TP]]).astype(int)

new_confusion_matrix = calculate_new_confusion_matrix(precision, recall, purchase_rate, customers)

new_TP = new_confusion_matrix[1,1]
new_FP = new_confusion_matrix[0,1]
new_FN = new_confusion_matrix[1,0]
new_TN = new_confusion_matrix[0,0]

print("Confusion Matrix:")
print(new_confusion_matrix)
```

```python
model_cost = advertisement_cost * (new_TP+new_FP)
model_lost = average_purchase * new_FN
model_gain = average_purchase * new_TP
model_value = model_gain - model_lost - model_cost
model_revenue = model_gain - model_cost
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
X = df_fin.drop(['Potential_Customer', 'Cust_Last_Purchase'], axis=1)
Y = df_fin['Potential_Customer']
```

```python
X.shape, Y.shape
```

```python
# Split numerical and categorical columns
NumCols = X.select_dtypes(include=['float']).columns.tolist()
CatCols = list(set(X.columns)-set(NumCols))

print(NumCols, CatCols)
```

```python
# One-hot encode categorical columns
X = pd.get_dummies(X, columns=CatCols, drop_first=True)
```

```python
# Standardize numerical columns
scaler = StandardScaler()
X[NumCols] = scaler.fit_transform(X[NumCols])
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.25,
                                                    random_state=42,
                                                    stratify=Y)
```

```python
# Convert target to numpy array
y_train = np.array(y_train)
y_test = np.array(y_test)
X_train.shape
```

```python
import tensorflow as tf

# Define the neural network
model = Sequential([
    Dense(14, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.4),  # Increased dropout for regularization
    Dense(32, activation='relu'),
    Dropout(0.4),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')  # Sigmoid activation for binary classification
])

def focal_loss(alpha=0.25, gamma=2):
    def loss(y_true, y_pred):
        bce = BinaryCrossentropy()(y_true, y_pred)
        pt = tf.exp(-bce)
        focal_loss = alpha * (1 - pt) ** gamma * bce
        return focal_loss
    return loss

# Compile the model with class weights
model.compile(optimizer='adam',
              loss=focal_loss(),  # Call the focal_loss function to get the actual loss function
              metrics=['Accuracy','Recall'])

# Add EarlyStopping to avoid overfitting
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=100,
                    batch_size=32,
                    class_weight={0: 1, 1: 1},  # Give higher weight to the positive class
                    callbacks=[early_stop])
```

```python
# Evaluate on the test set
test_loss, test_accuracy, test_recall = model.evaluate(X_test, y_test)

print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
print(f"Test Recall: {test_recall}")

# Predictions
threshold = 0.44  # Lower the threshold
y_pred = (model.predict(X_test) > threshold).astype("int32")

# Confusion matrix and classification report
from sklearn.metrics import classification_report, confusion_matrix

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
